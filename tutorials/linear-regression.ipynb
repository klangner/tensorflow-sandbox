# Linear regression

Create noisy linear regression model with known parameters.
Try to find good estimator for it based on data generated from the model.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

Define model parameters

```python
p0 = -8.9
p1 = 3.4
noise = 9.2
```

Create training set

```python
train_size = 1000
x_train = np.linspace(-10., 10., train_size)
y_train = p1 * x_train + p0 + np.random.randn(train_size) * noise
```

Plot training data

```python
plt.scatter(x_train, y_train)
plt.show()
```

Use TF to estimate best line to fit given data.

First create graph:

```python
learning_rate = 0.01
# Placeholders
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
# Model
W = tf.Variable(1.1, tf.float32)
b = tf.Variable(1.1, tf.float32)
linear_model = W * x + b
# Loss function
squared_delta = tf.square(linear_model - y)
loss_op = tf.reduce_sum(squared_delta) / train_size
# Optimizer
train_op = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss_op)
```

And now train estimator

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(1000):
        sess.run(train_op, {x:x_train, y:y_train})
        if i % 100 == 0:
            print('loss: %f' % sess.run(loss_op, {x:x_train, y:y_train}))
    # Expected 3.4 -8.9
    print(sess.run([W, b]))
```
